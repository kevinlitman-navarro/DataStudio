```python
df = pd.read_csv('petfinderNY_database.csv')
```

```python
df.breeds = df.breeds.str.replace('[', '')
df.breeds = df.breeds.str.replace(']', '')
df.breeds = df.breeds.str.replace('\'', '')
```

```python
def make_a_list(row):
    row.breeds = [row.breeds]
    return row
```

```python
df.apply(make_a_list, axis=1)
```

```python
def expand_breeds(row):
    test_string = ','
    new_row = {}
    new_rows = []
    if test_string in row.breeds:
        breed_list = row.breeds.split(',')
        for pet_breed in breed_list[1:]:
            new_row['breeds'] = pet_breed 
            new_row['additional_info'] = row.additional_info
            new_row['age'] = row.age
            new_row['animal'] = row.animal
            new_row['description'] = row.description
            new_row['id'] = row.id
            new_row['lastUpdate'] = row.lastUpdate
            new_row['mixed'] = row.mixed
            new_row['name'] = row.name
            new_row['pet_size'] = row.pet_size
            new_row['picture_links'] = row.picture_links
            new_row['sex'] = row.sex
            new_row['shelterID'] = row.shelterID
            new_row['zipcode'] = row.zipcode
            new_row['altered'] = row.altered
            new_row['vaccinated'] = row.vaccinated
            new_row['housetrained'] = row.housetrained
            new_row['cynophobic'] = row.cynophobic
            new_row['hates_children'] = row.hates_children
            new_row['felinephobic'] = row.felinephobic
            new_row['specialNeeds'] = row.specialNeeds
            new_rows.append(new_row)
            #new_df = pd.DataFrame(new_rows)
    return(new_rows)
        #df1 = pd.DataFrame(new_rows)
```

```python
df1 = df.apply(expand_breeds, axis=1)
new_list = df1.get_values().tolist()
new_rows = []
for listing in new_list:
    try:
        if type(listing[0]) == dict:
            new_rows.append(listing)
    except:
        pass
new_rows

new_new_rows = []
for item in new_rows:
    new_new_rows.append(item[0])
    
new_new_rows
```

```python
df_long = df.append(new_new_rows)
```
