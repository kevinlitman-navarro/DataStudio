# I'm going to scrape a bunch of metadata on the tabs at Ultimate Guitar dot com to determine the best guitar songs for casual players/people trying to impress love interests

```python
from bs4 import BeautifulSoup
import requests
import html5lib
import time
import json
import datetime
import pandas as pd
```

# This is me figuring out how to scrape my data

```python
raw_html = requests.get('https://www.ultimate-guitar.com/top/tabs?order=rating_desc').content
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

The html is wrapped in scripts, but with some string slicing I can pull out a string that can be turned into a dictionary with json

```python
listings = soup_doc.find('section', class_='_3-q_x')
```

```python
listings_string = soup_doc.find_all('script')[9].text[31:-2]
```

```python
listings_dict = json.loads(listings_string)
```

```python
tab_list = listings_dict['data']['tabs']
```

```python
hits_list = listings_dict['data']['hits']
```

```python
tab_list[0]
```

I can put this straight into a dataframe!

```python
df = pd.DataFrame(tab_list)
```

```python
df[df.id == 1433]
```

```python
df_hits = pd.DataFrame(hits_list)
```

```python
df_hits
```

```python
df_hits.id = df_hits.id.astype(int)
```

```python
df_hits.dtypes
```

```python
df = pd.merge(df, df_hits, on='id')
```

```python
df
```

# Saving my df of the top 100 to a csv for later

```python
df.to_csv('top_100_rated_tabs_all_time.csv', index=False)
```

```python
df.sort_values(by='hits')
```

```python
df.date = df.date.astype(int)
```

```python
def to_datetime(row):
    row['datetime'] = datetime.datetime.fromtimestamp(row.date).strftime('%m-%d-%Y')
    return row
```

```python
df = df.apply(to_datetime, axis=1)
```

```python
df['datetime'] = pd.to_datetime(df.datetime, format='%m-%d-%Y')
```

```python
df.head()
```

```python
df.to_csv('top_100_rated_tabs_all_time.csv', index=False)
```

# Moving on... I'm going to look at individual artists

Starting with the beatles

```python
beatles_url = 'https://www.ultimate-guitar.com/artist/the_beatles_1916?page='
```

```python
beatles_url = 'https://www.ultimate-guitar.com/artist/the_beatles_1916?page='
page_num = 1
beatles_tab_list = []

for i in range(1,43):
    url = beatles_url + str(i)
    try:
        raw_html = requests.get(url).content
        soup_doc = BeautifulSoup(raw_html, "html.parser")
        listings_string = soup_doc.find_all('script')[9].text[31:-2]
        listings_dict = json.loads(listings_string)
        try:
            tab_list = listings_dict['data']['album_tabs']
        except:
            pass
        try:
            other_tab_list = listings_dict['data']['other_tabs']
        except:
            pass
        try:
            for tab in tab_list:
                beatles_tab_list.append(tab)
        except:
            pass
        try:
            for tab in other_tab_list:
                beatles_tab_list.append(tab)
        except:
            pass
        print('did it for page' + str(i))
    except:
        pass

    
df_beatles = pd.DataFrame(beatles_tab_list)
```

```python
df_beatles.dtypes
```

```python
df_beatles.song_name.value_counts()
```

```python
df_beatles.to_csv('beatles-tabs.csv', index=False)
```
